## 주피터 노트북 단축키
- enter : 편집상태로 변환

-  Shift + Enter : 셀이 실행이 되고 커서가 다음셀로 이동

- Ctrl + Enter : 셀이 실행되고 커서가 그 셀에 유지

- a : 위에 셀 한 개 추가

- b : 아래 셀 한 개 추가

- dd : 셀 삭제

- m : 셀을 마크다운 셀로 변환

## 판다스
- 대괄호 1개 [] : Series
- 대괄호 2개 [[]] : DataFrame
- 2개 이상을 불러올 때 DataFrame으로 불러와야 함
- 빈도수 구하기: df[col].value_counts()
- 정렬 (col 기준): df[col].sort_values()
- 정렬 (전체에서 val기준): df.sort_values(val)
- Drop:
    - col drop: df.drop([col], axis = 1)
    - index drop: df.drop([idx], axis = 0)
- groupby: df.groupby(['idx', 'col'])[val]
- pivot_table: df.pivot(index='idx', columns='col', values='val')
- 그래프 그리기: df.plot(), df.plot.bar(), df.plot.density()
